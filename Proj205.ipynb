```python
import pandas as pd
import numpy as np
```

```python
bikehr=pd.read_csv('hour.csv')
bikedy=pd.read_csv('day.csv')
```

```python
weekday_list = []
for i in range(7):
    weekd = bikehr.groupby('weekday').get_group(i)
    weekday_list.append(weekd)
sun = weekday_list[0]['cnt'].sum()/3292679
mon = weekday_list[1]['cnt'].sum()/3292679
tue = weekday_list[2]['cnt'].sum()/3292679
wed = weekday_list[3]['cnt'].sum()/3292679
thu = weekday_list[4]['cnt'].sum()/3292679
fri = weekday_list[5]['cnt'].sum()/3292679
sat = weekday_list[6]['cnt'].sum()/3292679
day = [[sun,mon,tue,wed,thu,fri,sat]]
```

```python
workingday_list = []
for i in range(2):
    workd = bikehr.groupby('workingday').get_group(i)
    workingday_list.append(workd)
noworking=workingday_list[0]['cnt'].sum()/3292679
working=workingday_list[1]['cnt'].sum()/3292679
workingday=[[noworking,working]]
workingday
```

```python
weathersit = []
for i in range(1,5):
    wts = bikehr.groupby('weathersit').get_group(i)
    weathersit.append(wts)
clear=weathersit[0]['cnt'].sum()/3292679
mist=weathersit[1]['cnt'].sum()/3292679
light=weathersit[2]['cnt'].sum()/3292679
heavy=weathersit[3]['cnt'].sum()/3292679
weather=[[clear,mist,light,heavy]]
weather
```

```python
bikedy= bikedy.replace({'weathersit': {0:0.7101126468750826 }})
bikedy= bikedy.replace({'weathersit': {1:0.24173385866037958 }})
bikedy= bikedy.replace({'weathersit': {2:0.04808576845784238 }})
bikedy= bikedy.replace({'weathersit': {3:6.772600669545984e-05 }})
```

```python
bikedy= bikedy.replace({'weekday': {0.000000: 0.1348528052689011}})
bikedy = bikedy.replace({'weekday': {1.000000: 0.13833811312915714}})
bikedy = bikedy.replace({'weekday': {2.000000: 0.1424703106497779}})
bikedy = bikedy.replace({'weekday': {3.000000: 0.14366660096535375}})
bikedy = bikedy.replace({'weekday': {4.000000: 0.14741643506700775}})
bikedy = bikedy.replace({'weekday': {5.000000: 0.14814380630483567}})
bikedy = bikedy.replace({'weekday': {6.000000: 0.14511192861496672}})
```

```python
bikedy= bikedy.replace({'workingday': {0: 0.30378576229264986}})
bikedy = bikedy.replace({'workingday': {1: 0.6962142377073501}})
```

```python
bikedy
```

```python
x=bikedy[['temp','windspeed','hum','weekday','weathersit','workingday']].to_numpy()
y=bikedy['cnt'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python
clf.fit(x_train,y_train)
clf.coef_
```

```python
clf.predict(x_test)
```

```python
y_pred = clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(y_pred)
plt.plot(y_test)
plt.xlabel('sample')
plt.ylabel('people')
plt.legend(['Pred', 'Test'])
```

```python
import numpy as np
idx = np.argsort(y_test)
plt.plot(y_pred[idx])
plt.plot(y_test[idx])
plt.xlabel('Samples')
plt.ylabel('Value (k$)')
plt.legend(['Pred', 'Test'])
```

```python
from sklearn.metrics import mean_squared_error
score = mean_squared_error(y_test, y_pred)
score
```

```python
from sklearn.model_selection import cross_val_score
scores = -cross_val_score(clf, 
                          x_train, 
                          y_train, 
                          cv=5, 
                          scoring='neg_mean_squared_error')
print(scores)
```

```python
import seaborn as sns
ax = sns.barplot(x=list(range(5)), y=scores)
ax.set(xlabel='Folds', ylabel='MSE')
ax.set_title('MSE: {:0.2f} +/- {:0.2f}'.format(scores.mean(), scores.std()*2))
plt.show()
```
